```python
from sympy import isprime
import matplotlib.pyplot as plt
def kaibun(str): return 1 if str == str[::-1] else 0
x,y,num=0,-1,10000000
sosu_kaibun=[]
normal_kaibun=[]
sosu=[]
while x<num:
  if isprime(x):
    sosu.append(x)
  if kaibun(str(x)):
    normal_kaibun.append(x)
  if isprime(x) and kaibun(str(x)):
    sosu_kaibun.append(x)
  x+=1
print("回文普通",normal_kaibun)
print("回文素数",sosu_kaibun)
print("普通素数",sosu)
kaidan = []
up = 1
for nk in normal_kaibun:
  if isprime(nk)!=True:
    kaidan.append(up)
  if isprime(nk):
    up+=1
    kaidan.append(up)
print(kaidan)
plt.figure(figsize=(1200,720))
plt.figure(dpi=1200)
plt.title("Kaibun-Research-kaidan({})".format(num))
Y=range(len(normal_kaibun))
plt.plot(Y,kaidan)
plt.show()
```
