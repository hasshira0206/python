```python
def ret(inp):
    from sympy import isprime
    x=3
    li=[]
    num_list=[]
    while x<=inp:#ここいじれ
        num_list=[x,x-1,x-2]
        num=num_list[0]*num_list[1]*num_list[2]
        num/=6
        li.append(num)
        x+=1
    li2=[]
    li3=[]
    for x in range(len(li)-1):
        li2.append(li[x+1]-li[x])
    x=0
    while x<len(li2)-1:
        X=li2[x+1]-li2[x]
        li3.append(X)
        x+=2
    con=len(li3)
    x=0
    for l in li3:
        if isprime(int(l)):
            x+=1
    return x,con
```

```python
lis=[]
for x in range(3,1000):#ここをいじれ
    num=ret(x)
    if num[1]!=0:
        lis.append(num[0]/num[1]*100)
print(lis)
```

```python
import matplotlib.pyplot as plt
plt.plot(lis)
plt.show()
```
