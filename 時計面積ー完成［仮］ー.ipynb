```python
import matplotlib.pyplot as plt
import sympy
minute=[]
hour=[]
kakudo=[]
x = sympy.Symbol('x')
y = sympy.Symbol('y')
t = sympy.Symbol('t')
t=0
while t<12:
    if t>=6:
        t-=1
    expr1 = y-0.5*x-(30*t)
    #print(expr1)
    expr2 = y-6*x+180

    #print(sympy.solve([expr1, expr2]))
    n=sympy.solve([expr1,expr2])
    n=n[x]
    if n>59:
        while True:
            n-=60
            t+=1
            if n<60:
                break
    hour.append(t)
    minute.append(n)
    t+=1
for h,m in zip(hour,minute):
    kakudo=round(m)*6
    print(h,round(m),kakudo)
kari=0
y,x,tan,tyo,num,x1,y1,li=0,0,0,0,0,0,0,[]
def pri(T,x1,y1):
    return x,y,round(T),x1,y1
while True:
    if y1>59:
        y,y1=0,0
        x1+=1
        kari+=1
    if x1>=12:
        break
    if kari==5:
        kari-=1
    N=minute[kari]
    a=0
    if y1<N and y>x:
        a=y-x
        T=3.14*10**2*a/360
        print(pri(T,x1,y1))
        li.append(T)
    if y1<N and y<x:
        a=x-y
        T=3.14*10**2*a/360
        print(pri(T,x1,y1))
        li.append(T)
    if y1==N:
        a=180
        T=3.14*10**2*a/360
        print(pri(T,x1,y1))
        li.append(T)
    if y1>N and y>x:
        a=x+(360-y)
        T=3.14*10**2*a/360
        print(pri(T,x1,y1))
        li.append(T)
    if y1>N and y<x:
        a=x+(360-x)+t
        T=3.14*10**2*a/360
        print(pri(T,x1,y1))
        li.append(T)
    x+=0.5
    y+=6
    y1+=1
print(li)
plt.figure(figsize=(1200,720))
plt.figure(dpi=1200)
plt.title("12H of Clock's Fan-shaped area")
plt.plot(li)
plt.show()
```
